```python
import bm_utils

X_train, X_test, Y_train, Y_test = bm_utils.load_dataset("../../preprocessing/6_less_tensor.npz", test_size=0.1, random_state=42)
print(len(X_train), len(X_test))
```

```python
# benchmark to use
import chess
import random
random.seed(42)

# randomly picks 5 legal moves for a given board
def random_predict(board):
    legal_moves = list(board.legal_moves)
    top5_moves = []
    if len(legal_moves) <= 5:
        top5_moves = legal_moves
    else:
        top5_moves = random.sample(legal_moves, 5)
    
    return top5_moves
```

```python
bm_utils.evaluate(X_test, Y_test, random_predict)
```
